# 1.   Pengecekan data set


```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_csv("/content/SMDR.JK(01012020_ 28012023).csv")
print(data.head())
```

#2.   Cleansing data jika ada null values


```python
data.isnull().values.sum()
```

```python
data = data.dropna()
```

```python
data.isnull().values.sum()
```

#3.   Visualisasi grafik


```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "SAMUDRA INDONESIA Stock Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

```python
print(data.corr())
```

#4.   Menggunakan library Autots


```python
pip install autots
```

#5.   Proses prediksi


```python
from autots import AutoTS
model = AutoTS(forecast_length=3, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
```

#6.   Hasil prediksi


```python
print(forecast)
```

Harga prediksi yang ditemukan adalah

Closing saham SMDR pada tanggal 30 Januari sampai 01 Februari adalah :
